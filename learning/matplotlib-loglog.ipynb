```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def power_law(x, a, b):
    return a * (x**b)

# Generate x values (must be positive for log scale)
x = np.logspace(0.1, 3, 100) # Generates points spaced logarithmically

# Define parameters for the power-law function
a_val = 5
b_val = 2

# Calculate y values
y = power_law(x, a_val, b_val)

# Create the log-log plot
plt.figure(figsize=(8, 6)) # Optional: set figure size
plt.loglog(x, y, 'b-', label=f'y = {a_val} * x^{b_val}') # 'b-' for blue line

# Add labels and title
plt.xlabel('X-axis (log scale)')
plt.ylabel('Y-axis (log scale)')
plt.title('Log-Log Plot of a Power-Law Function')
plt.grid(True, which="both", ls="-", color='0.7') # Add grid for better readability
plt.legend() # Display the legend

# Show the plot
plt.show()
```
